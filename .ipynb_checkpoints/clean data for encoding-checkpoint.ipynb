```python
# copy dataset
health_dataset_1=health_dataset.copy()
health_dataset_2=health_dataset.copy() 
health_dataset_3=health_dataset.copy() 
health_dataset_4=health_dataset.copy()
health_dataset_5=health_dataset.copy() 
health_dataset_6=health_dataset.copy()  
health_dataset_7=health_dataset.copy()
health_dataset_8=health_dataset.copy() 
health_dataset_9=health_dataset.copy()  
health_dataset_10=health_dataset.copy() 




```

```python
#check duplicated raw  
print(health_dataset_1.duplicated()) 
```

```python
# drop duplicated raw  
health_dataset_1.drop_duplicates(inplace=True)
```

```python
print(health_dataset_1.to_string())
```

```python
# convert empty cells and spaces into NaN
health_dataset_1 = health_dataset_1.replace("", np.nan)
health_dataset_1 = health_dataset_1.replace(" ", np.nan)
```

```python
# numerical column null value replace by mean /median

health_dataset_1["Age"] = health_dataset_1["Age"].fillna(health_dataset_1["Age"].median())


health_dataset_1["BMI"] = health_dataset_1["BMI"].fillna(health_dataset_1["BMI"].mean())


health_dataset_1["Heart_Rate"] = health_dataset_1["Heart_Rate"].fillna(health_dataset_1["Heart_Rate"].mean())


health_dataset_1["Sleep_Hours"] = health_dataset_1["Sleep_Hours"].fillna(health_dataset_1["Sleep_Hours"].mean())


health_dataset_1["Blood_Sugar_Level"] = health_dataset_1["Blood_Sugar_Level"].fillna(health_dataset_1["Blood_Sugar_Level"].median())
```

```python
# categorial column null value replace by mode

health_dataset_1["Gender"] = health_dataset_1["Gender"].fillna(health_dataset_1["Gender"].mode()[0])

health_dataset_1["Smoking"] = health_dataset_1["Smoking"].fillna(health_dataset_1["Smoking"].mode()[0])

health_dataset_1["Alcohol_Consumption"] = health_dataset_1["Alcohol_Consumption"].fillna(health_dataset_1["Alcohol_Consumption"].mode()[0])

health_dataset_1["Exercise_Frequency"] = health_dataset_1["Exercise_Frequency"].fillna(health_dataset_1["Exercise_Frequency"].mode()[0])

health_dataset_1["Blood_Pressure"] = health_dataset_1["Blood_Pressure"].fillna(health_dataset_1["Blood_Pressure"].mode()[0])

health_dataset_1["Cholesterol_Level"] = health_dataset_1["Cholesterol_Level"].fillna(health_dataset_1["Cholesterol_Level"].mode()[0])

health_dataset_1["Stress_Level"] = health_dataset_1["Stress_Level"].fillna(health_dataset_1["Stress_Level"].mode()[0])

health_dataset_1["Medication_Use"] = health_dataset_1["Medication_Use"].fillna(health_dataset_1["Medication_Use"].mode()[0])

health_dataset_1["Family_History"] = health_dataset_1["Family_History"].fillna(health_dataset_1["Family_History"].mode()[0])
```

```python
#check again have any null value ar not 
health_dataset_1.isna().sum()
```

```python
print(health_dataset_1.to_string())
```

```python
#now my dataset is ready for encoding 
print(health_dataset_1.info())
```
