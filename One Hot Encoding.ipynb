```python
#using loop for fit and transform 
#encoding is not applicable for numeric data 

from pandas.api.types import is_numeric_dtype   #import method for check numeric or not 
for col in clean_health_dataset_1.columns: 
    if is_numeric_dtype(clean_health_dataset_1[col]):   #is_numeric_dtype ekti method  numeric kina check korar
        continue
    else: 
        new_dummy=pd.get_dummies(clean_health_dataset_1[col],prefix=col).astype('float64')
        clean_health_dataset_1.drop([col],axis=1,inplace=True)
        clean_health_dataset_1=pd.concat([clean_health_dataset_1,new_dummy],axis=1)
```
